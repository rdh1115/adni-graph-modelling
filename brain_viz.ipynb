```python
def plot_trajectory_per_cluster(cluster_list, fp_name, norm_node=True):
    cluster_arr = np.array(cluster_list)  # shape [time steps, nodes (68)]
    
    if 'mri' in fp_name:
        shape = cluster_arr.shape
        cluster_arr = cluster_arr.reshape(shape[0], 68, -1)[:, :, 0]
        cluster_arr = cluster_arr.reshape(shape[0], 68)
        print(f"reshaped from {shape} to {cluster_arr.shape}")
    if norm_node:
        # Z-score normalize across time for each node
        normalized_data = []
        for node_idx in range(cluster_arr.shape[1]):  # Iterate over nodes
            node_data = cluster_arr[:, node_idx]  # Extract time series for a node
            mean = np.mean(node_data)
            std = np.std(node_data)
            normalized = (node_data - mean) / std
            normalized_data.append(normalized)
        # Transpose normalized_data to match the required shape [time steps, nodes]
        normalized_data = np.array(normalized_data).T
    else:
        # Z-score normalize across all dimensions (time and nodes)
        mean = np.mean(cluster_arr)
        std = np.std(cluster_arr)
        normalized_data = (cluster_arr - mean) / std
    
    # Generate cortical plots
    plt = plot_cortical(
        array_name=[parcel_to_surface(arr, 'aparc_fsa5') for arr in normalized_data], 
        surface_name="fsa5", 
        size=(4000, 2000), 
        interactive=True, 
        filename=fp_name,
        embed_nb=True,
        screenshot=True,
        color_bar='bottom',
        label_text={
            'top': ['left lateral', 'left medial', 'right medial', 'right lateral'],
            'left': [f'visit {i}' for i in range(normalized_data.shape[0])]
        },
        transparent_bg=False,
        share=True,
    )
    return

def plot_trajectory(cluster_list, fp_name, norm_node=True, top_k=5):
    cluster_arr = np.array(cluster_list)  # shape [num_clusters, num_time, num_nodes]
    
    if 'mri' in fp_name:
        shape = cluster_arr.shape
        cluster_arr = cluster_arr.reshape(shape[0], shape[1], 68, -1)[..., 0]
        print(f"reshaped from {shape} to {cluster_arr.shape}")
    
    if norm_node:
        # Z-score normalize across clusters for each node
        normalized_data = []
        for node_idx in range(cluster_arr.shape[2]):  # Iterate over nodes
            node_data = cluster_arr[:, :, node_idx]  # Extract data across clusters and time for a node
            mean = np.mean(node_data)
            std = np.std(node_data)
            normalized = (node_data - mean) / std
            normalized_data.append(normalized)
        
        # Transpose normalized_data to match the required shape [num_clusters, num_time, num_nodes]
        normalized_data = np.transpose(normalized_data, (1, 2, 0))
    else:
        # Z-score normalize across all dimensions (clusters, time, nodes)
        mean = np.mean(cluster_arr)
        std = np.std(cluster_arr)
        normalized_data = (cluster_arr - mean) / std

    # Find nodes with the top k most positive and negative cumulative changes
    for cluster_idx in range(normalized_data.shape[0]):  # Iterate over clusters
        cluster_normalized = normalized_data[cluster_idx, :, :]  # [time steps, nodes]
        changes = np.sum(np.diff(cluster_normalized, axis=0), axis=0)  # Cumulative change across all time steps
        
        # Sort changes
        top_k_positive_indices = np.argsort(changes)[-top_k:][::-1]  # Indices of top k positive changes
        top_k_negative_indices = np.argsort(changes)[:top_k]         # Indices of top k negative changes
        
        print(f"Cluster {cluster_idx}:")
        print(f"  Top {top_k} nodes with positive cumulative changes:")
        for idx in top_k_positive_indices:
            print(f"    Node {idx}, {NODE_LABELS[idx]}: Change {changes[idx]:.3f}")
        
        print(f"  Top {top_k} nodes with negative cumulative changes:")
        for idx in top_k_negative_indices:
            print(f"    Node {idx}, {NODE_LABELS[idx]}: Change {changes[idx]:.3f}")
    
    # Prepare cortical plots
    cortical_data = []
    for cluster_idx in range(normalized_data.shape[0]):  # Iterate over clusters
        cluster_normalized = normalized_data[cluster_idx, :, :]  # [time steps, nodes]
        cortical_data.append([parcel_to_surface(arr, 'aparc_fsa5') for arr in cluster_normalized])  # normalize each time step

    # Generate cortical plots for each cluster
    for cluster_idx, arr in enumerate(cortical_data):
        plot_cortical(
            array_name=arr, 
            surface_name="fsa5", 
            size=(4000, 2000), 
            interactive=True, 
            filename=f"{fp_name}_cluster{cluster_idx}_{norm_node}.png",
            embed_nb=True, 
            screenshot=True, 
            color_bar='bottom',
            label_text={
                'top': ['left lateral', 'left medial', 'right medial', 'right lateral'],
                'left': [f'visit {i}' for i in range(cluster_arr.shape[1])]
            },
            transparent_bg=False,
            share=True,
        )
    return
```
